```python
import numpy as np
import cv2
from glob import glob
import matplotlib.pyplot as plt
```

```python
fpath = glob('./data/female/*.jpg')
mpath = glob('./data/male/*.jpg')
```

```python
len(fpath)
```

```python
#Read Image and convert to RGB
img = cv2.imread(fpath[1]) #read image in BGR format
img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

#apply Haar Cascade Classifier
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
gray = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
faces_list = haar.detectMultiScale(gray,1.5,5)
print(faces_list)

for x,y,w,h in faces_list: #there might be multiple face so we are using for loop
    cv2.rectangle(img_rgb,(x,y),(x+w,y+h),(0,255,0),2)
    #crop face
    roi = img_rgb[y:y+h,x:x+h]
    plt.imshow(roi)
    plt.axis('off')
    plt.show()

#save Image

```

## crop all female faces

```python
haar = cv2.CascadeClassifier('./model/haarcascade_frontalface_default.xml')
```

```python
for i in range(len(fpath)):
    try:
        #Read Image and convert to RGB
        img = cv2.imread(fpath[i]) #read image in BGR format
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

        #apply Haar Cascade Classifier
        gray = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)

        for x,y,w,h in faces_list: #there might be multiple face so we are using for loop
            #crop face
            roi = img[y:y+h,x:x+h]
            #save image
            cv2.imwrite(f'./crop_data/female/female_{i}.jpg',roi)
            print('Image Successfully processed')
    except:
        print('unable to process')
```

## Crop all male Images

```python
for i in range(len(mpath)):
    try:
        #Read Image and convert to RGB
        img = cv2.imread(mpath[i]) #read image in BGR format
        img_rgb = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)

        #apply Haar Cascade Classifier
        gray = cv2.cvtColor(img_rgb,cv2.COLOR_RGB2GRAY)
        faces_list = haar.detectMultiScale(gray,1.5,5)

        for x,y,w,h in faces_list: #there might be multiple face so we are using for loop
            #crop face
            roi = img[y:y+h,x:x+h]
            #save image
            cv2.imwrite(f'./crop_data/male/male_{i}.jpg',roi)
            print('Image Successfully processed')
    except:
        print('unable to process')
```
